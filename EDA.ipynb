## 1: Wczytanie i ogólny opis danych

Import bibliotek i wczytanie danych

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("water_potability.csv")
```

```python
df.info() #typy danych w kolumnach
```

## 2: Analiza zmiennych numerycznych

Statystyki opisowe

```python
df.describe()
```

Wizualizacje rozkładów

```python
df.hist(figsize=(15, 10), bins=30)
plt.tight_layout()
plt.show()
```

Boxploty (outliery)

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

## 3: Analiza zmiennych kategorycznych

Zmienna docelowa


```python
df["is_safe"].value_counts()
df["is_safe"].value_counts(normalize=True) * 100 #procentami
```

```python
sns.countplot(x="is_safe", data=df)
plt.title("Rozkład zmiennej is_safe")
plt.xlabel("Czy woda jest bezpieczna (0 = nie, 1 = tak)")
plt.ylabel("Liczba próbek")
plt.show()
```

## 4: Analiza zależności między zmiennymi

Macierz korelacji

```python
df = df.replace('#NUM!', np.nan) #Zaznaczam braki, zamieniam „#NUM!” na NaN
df = df.apply(pd.to_numeric, errors='coerce') #Przyprowadzam kolumny do liczby (0 lub 1)
corr = df.corr()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.title("Macierz korelacji")
plt.show()
```

Wybrane zależności

```python
sns.scatterplot(
    x="ammonia",
    y="nitrates",
    hue="is_safe",
    data=df
)
plt.title("Zależność amoniaku i azotanów a jakość wody")
plt.show()
```

W zbiorze danych brakuje wspólnej zmiennej dotyczącej jakości wody. W związku z tym do analizy zależności między parametrami wykorzystałem inne właściwości chemiczne, na przykład stężenie amoniaku i azotanów.

Braki danych

#### Decyzja

W zbiorze danych występują braki spowodowane obecnością wartości #NUM!, które zostały zastąpione wartością NaN. Na etapie EDA braki w danych nie zostały jeszcze uzupełnione, aby uniknąć nieporozumień podczas pracy z modelowaniem. Uzupełnienie braków zostanie wykonane na etapie przygotowania danych do modelowania.

#### Outliery

Na wykresach typu box-plot widoczne są odchylenia wartości wielu charakterystyk. Na etapie EDA postanowiłem ich nie usuwać, ponieważ mogą one stanowić rzeczywiste, choć rzadkie, przypadki jakości wody.
